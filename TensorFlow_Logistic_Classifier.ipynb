```python
import tensorflow as tf
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
import math
def get_weights(n_features,n_labels):
    # Return Tensorflow weight
    weights = tf.Variable(tf.random_normal([n_features,n_labels]))
    return weights

def get_biases(n_labels):
    # Return Tensorflow bias
    bias = tf.Variable(tf.random_normal([n_labels])) 
    return bias
def linear_function(input,w,b):
    return tf.add(tf.matmul(input,w),b)
def batches(batch_size,features,labels):
    assert len(features)==len(labels)
#     Create batches of features and labels
    out_batches = []
    sample_size=len(features)
   
    for start_i in range(0,sample_size,batch_size):
        end_i=start_i+batch_size
        batch=[features[start_i:end_i],labels[start_i:end_i]]
        out_batches.append(batch)
   
    return out_batches
        
def print_epoch_stats(epoch_i, sess, last_features, last_labels):
    #     Print cost and validation accuracy of an epoch
    current_cost=sess.run(cost,feed_dict={features:last_features,labels:last_labels})
    valid_accuracy=sess.run(accuracy,feed_dict={features:last_features,labels:last_labels})
    print('Epoch: {:<4} - Cost: {:<8.3} Valid Accuracy: {:<5.3}'.format(
        epoch_i,
        current_cost,
        valid_accuracy))

    
    
    
```

```python

#Define a function to load feature and lables from data sets
def mnist_features_labels():
   
   
    #Download MNIST data set if not present in local
    mnist= input_data.read_data_sets('datasets/mnist', one_hot=True)
    #Extract features and lables from the dataset
    
    train_features=mnist.train.images
    test_features = mnist.test.images
    
    train_labels = mnist.train.labels.astype(np.float32)
    test_labels = mnist.test.labels.astype(np.float32)
    return train_features,train_labels,test_features,test_labels
    




    
```

```python
# Tranning the data
# train_features,train_labels = mnist_features_labels(n_classes)
train_features,train_labels,test_features,test_labels = mnist_features_labels()

   
```

```python
# number of the features of 28x28 image is = 784
n_features=784
# number of lables 
n_classes = 10

#Define the learning rate
learning_rate=0.01

# Place holder for features and lables
features = tf.placeholder(tf.float32,[None,n_features])
labels = tf.placeholder(tf.float32,[None,n_classes])

# Weight and Biases 
w = get_weights(n_features,n_classes)
b = get_biases(n_classes)

# Calculte the logits. Linear Model xw+b
logits = linear_function(features,w,b)
print(logits)

```

```python



# Define loss and optimizer
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=labels))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)


# Calculte the accuracy 
correct_prediction = tf.equal(tf.argmax(logits,1), tf.argmax(labels, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))

    
# Batch size
batch_size=128
epochs =150
assert batch_size is not None, 'Must set the batch size'
train_batchs= batches(batch_size,train_features,train_labels)
init = tf.global_variables_initializer()
with tf.Session() as sess:
    # initialize the session variables
    sess.run(init)
    
    for epoch_i in range(epochs):
        # Run the optimizer and get the loss
        for batch_feature,batch_label in train_batchs:
            train_feed_dict={features:batch_feature,labels:batch_label}
            sess.run(optimizer,feed_dict=train_feed_dict)
        print_epoch_stats(epoch_i,sess,batch_feature,batch_label)
        
    #Calculate the accuracy for test data
    test_accuracy=sess.run(accuracy,feed_dict={features:test_features,labels:test_labels})
    
# Loss
print('Test Accuracy: {}'.format(test_accuracy))
    
   
    
    
    

 
```
